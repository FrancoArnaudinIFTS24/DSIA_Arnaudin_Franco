## Ejercicio 2 \- Dataset simple con Pandas

```
notas = {
    'estudiante': ['Ana', 'Juan', 'Carla', 'Pedro', 'Lucia'],
    'nota': [8, 6, 9, 5, 7]
}
df = pd.DataFrame(notas)

print(df)

print("Promedio de notas: ", df['nota'].mean() )

print("Nota máxima: ", df['nota'].max() )

print("Nota mínima: ", df['nota'].min() )
```
