### Ejercicio 2 \- Condicionales

```
age = int(input("Ingrese su edad: "))
if age >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

### Ejercicio 3 \- Listas y bucles

```
num_list = [1,5,87,-6,3]
acum = 0
for i in num_list:
    print(i)
    acum += i
avg = acum / len(num_list)
print("Suma de los valores: ", acum)
print("Promedio: ", avg)
```
